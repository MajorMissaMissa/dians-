```python
import requests
import pandas as pd
from datetime import datetime, timedelta
from tqdm import tqdm
import time
from google.colab import files  # Only needed if running in Google Colab

# Define the start and end dates for the data you want to download
start_date = datetime(2014, 1, 1)
end_date = datetime(2024, 10, 31)

# Initialize an empty list to store data
data_frames = []

# Use a session for efficient requests
session = requests.Session()

# Loop through each date and download the data
current_date = start_date
date_range = (end_date - start_date).days + 1

for _ in tqdm(range(date_range), desc="Downloading data"):
    # Format the date as YYYY-MM-DD
    date_str = current_date.strftime('%Y-%m-%d')

    # URL with date parameter (replace 'REPL' with the actual symbol if needed)
    url = f"https://www.mse.mk/daily/stats/symbolhistory/REPL?date={date_str}"

    # Retry logic in case of failed requests
    for attempt in range(3):
        try:
            response = session.get(url, timeout=10)
            if response.status_code == 200:
                # Assume data is in HTML format; parse the first table
                df = pd.read_html(response.content)[0]
                data_frames.append(df)
                print(f"Data for {date_str} downloaded successfully.")
                break  # Exit retry loop on success
            else:
                print(f"Failed to download data for {date_str}. Status code: {response.status_code}")
        except Exception as e:
            print(f"Error downloading data for {date_str}: {e}")
            time.sleep(1)  # Wait 1 second before retrying
    else:
        print(f"Failed to download data for {date_str} after 3 attempts.")

    # Move to the next day
    current_date += timedelta(days=1)

# Combine all data into a single DataFrame and save as CSV
if data_frames:
    all_data = pd.concat(data_frames, ignore_index=True)
    csv_file_path = "/content/historical_data.csv"
    all_data.to_csv(csv_file_path, index=False)
    print("Data saved to historical_data.csv")

    # Automatically download the file to your computer
    files.download(csv_file_path)
else:
    print("No data downloaded.")

```

```python
import requests
from datetime import datetime, timedelta
import pandas as pd

# Example: Fetch data by month (instead of a large date range)
start_date = datetime(2014, 1, 1)
end_date = datetime(2024, 10, 31)

current_date = start_date
all_data = []  # To store all fetched data

while current_date <= end_date:
    month_end = current_date.replace(day=28) + timedelta(days=4)  # Ensure we get the last day of the month
    month_end = month_end - timedelta(days=month_end.day)

    url = f'https://www.mse.mk/daily/stats/symbolhistory/REPL?date={current_date.strftime("%Y-%m-%d")}'
    response = requests.get(url, timeout=30)

    # Process response if successful
    if response.status_code == 200:
        print(f"Fetched data for {current_date.strftime('%Y-%m-%d')}")
        try:
            # Parse the first table in the HTML response
            df = pd.read_html(response.content)[0]

            # Replace '####' with None (missing value)
            df = df.applymap(lambda x: None if x == '####' else x)

            # Remove columns where all values are missing, '####', or only contain '0' or negative numbers
            df = df.dropna(axis=1, how='all')  # Remove columns where all values are missing
            df = df.loc[:, (df != 0).any(axis=0)]  # Remove columns where all values are '0'
            df = df.loc[:, (df >= 0).any(axis=0)]  # Remove columns with negative numbers only

            # Remove rows where any value is '0', negative, invalid name, or missing (None)
            df = df[~(df == 0).any(axis=1)]  # Remove rows with any '0' values
            df = df[~(df < 0).any(axis=1)]  # Remove rows with any negative numbers
            df = df.dropna(how='all')  # Remove rows where all values are missing
            df = df[~df.applymap(lambda x: isinstance(x, str)).any(axis=1)]  # Remove rows with invalid names (strings)

            # Add the DataFrame to the list if it has data
            if not df.empty:
                all_data.append(df)
            else:
                print(f"No valid data for {current_date.strftime('%Y-%m-%d')}")
        except Exception as e:
            print(f"Error processing data for {current_date.strftime('%Y-%m-%d')}: {e}")
    else:
        print(f"Failed to fetch data for {current_date.strftime('%Y-%m-%d')}")

    # Move to the next month
    current_date = month_end + timedelta(days=1)

# After the loop, merge all DataFrames and save to a CSV file
if all_data:
    merged_data = pd.concat(all_data, ignore_index=True)

    # Save to a CSV, only the fully populated data
    merged_data.to_csv('mse_data_cleaned_no_zeros_negatives_and_invalid_columns.csv', index=False)  # Save the cleaned data as CSV
    print("Data saved to mse_data_cleaned_no_zeros_negatives_and_invalid_columns.csv")
else:
    print("No data was fetched.")
```

```python
import requests
from datetime import datetime, timedelta
import pandas as pd

# Example: Fetch data by month (instead of a large date range)
start_date = datetime(2014, 1, 1)
end_date = datetime(2024, 10, 31)

current_date = start_date
all_data = []  # To store all fetched data

while current_date <= end_date:
    month_end = current_date.replace(day=28) + timedelta(days=4)  # Ensure we get the last day of the month
    month_end = month_end - timedelta(days=month_end.day)

    url = f'https://www.mse.mk/daily/stats/symbolhistory/REPL?date={current_date.strftime("%Y-%m-%d")}'
    response = requests.get(url, timeout=30)

    # Process response if successful
    if response.status_code == 200:
        print(f"Fetched data for {current_date.strftime('%Y-%m-%d')}")
        try:
            # Parse the first table in the HTML response
            df = pd.read_html(response.content)[0]

            # Replace '####' with None (missing value)
            df = df.applymap(lambda x: None if x == '####' else x)

            # Convert all columns to numeric, forcing errors to NaN (ignoring any non-numeric values)
            df = df.apply(pd.to_numeric, errors='coerce')

            # Remove columns where all values are missing, '####', or only contain '0' or negative numbers
            df = df.dropna(axis=1, how='all')  # Remove columns where all values are missing
            df = df.loc[:, (df != 0).any(axis=0)]  # Remove columns where all values are '0'
            df = df.loc[:, (df >= 0).any(axis=0)]  # Remove columns with negative numbers only

            # Remove rows where any value is '0', negative, invalid name, or missing (None)
            df = df[~(df == 0).any(axis=1)]  # Remove rows with any '0' values
            df = df[~(df < 0).any(axis=1)]  # Remove rows with any negative numbers
            df = df.dropna(how='all')  # Remove rows where all values are missing

            # Add the DataFrame to the list if it has data
            if not df.empty:
                all_data.append(df)
            else:
                print(f"No valid data for {current_date.strftime('%Y-%m-%d')}")
        except Exception as e:
            print(f"Error processing data for {current_date.strftime('%Y-%m-%d')}: {e}")
    else:
        print(f"Failed to fetch data for {current_date.strftime('%Y-%m-%d')}")

    # Move to the next month
    current_date = month_end + timedelta(days=1)

# After the loop, merge all DataFrames and save to a CSV file
if all_data:
    merged_data = pd.concat(all_data, ignore_index=True)

    # Save to a CSV, only the fully populated data
    merged_data.to_csv('mse_data_cleaned_no_zeros_negatives_and_invalid_columns.csv', index=False)  # Save the cleaned data as CSV
    print("Data saved to mse_data_cleaned_no_zeros_negatives_and_invalid_columns.csv")
else:
    print("No data was fetched.")
```
